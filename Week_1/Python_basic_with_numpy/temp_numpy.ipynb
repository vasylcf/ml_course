```python
def sigmoid(x):
    """
    Compute the sigmoid of x

    Arguments:
    x -- A scalar or numpy array of any size

    Return:
    s -- sigmoid(x)
    """
    
    ### START CODE HERE ### (≈ 1 line of code)
    s = 1/(1+np.exp(-x))
    ### END CODE HERE ###
    
    return s
```

```python
x=np.linspace(-10,10,100)
y=-(np.log(sigmoid(x))+np.log(1-sigmoid(x)))
```

```python
plt.plot(x,y)
plt.grid=1
```
